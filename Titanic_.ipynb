```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df= sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.countplot(x= 'alive', data= titanic_df)
plt.show()
```

```python
sns.countplot(x= 'alive', hue= 'who', data= titanic_df).set_title("Count of Survivors by Gender")
plt.show()
```

```python
two_columns_hue= titanic_df[['class', 'sex']].apply(tuple, axis=1)
two_columns_hue.unique()
```

```python
hue_order= [('First', 'female'), ('Second', 'female'), ('Third', 'female'),
            ('First', 'male'), ('Second', 'male'), ('Third', 'male')]
```

```python
sns.countplot(x= 'alive', hue= two_columns_hue, hue_order= hue_order, data= titanic_df)
plt.title('Count of Survivors by Class and Gender')
plt.legend(title= 'Class, Sex')
plt.show()
```

```python
sns.violinplot(x='who', y= 'age', data= titanic_df, hue= 'alive')
plt.title('Age Distribution by Passenger and Survival Status')
plt.show()
```

```python
ax= sns.boxplot(x= 'class', y= 'fare', data= titanic_df, hue= 'who')
ax.set_title('Fare Distribution by Class and Passenger Type')
plt.show()
```

```python
g= sns.FacetGrid(data= titanic_df, col= 'class', row= 'who')
g.map(sns.histplot, "age")
plt.show()
```

```python
taxis_df= sns.load_dataset('taxis')
taxis_df.head()
```

```python
sns.pairplot(taxis_df, hue= 'passengers')
plt.show()
```

```python
taxis_corr= taxis_df.select_dtypes('number').corr()
taxis_corr
```
